# Following Code  is using Librosa 

```python
import librosa
import librosa.display
import pickle 
import os
import numpy as np
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
y_, sr = librosa.load('test1-01.wav')
```

```python
try:
    mfcc = librosa.feature.mfcc(y=y_, sr=sr, n_mfcc=40)
except:
    print ('Skipping this file MFCCs Cant')
```

```python
mfcc.shape
```

```python
ig, ax = plt.subplots()
mfcc_data= np.swapaxes(mfcc, 0 ,1)
cax = ax.imshow(mfcc, interpolation='nearest', cmap=cm.coolwarm, origin='lower', aspect='auto')
ax.set_title('MFCC')
#Showing mfcc_data
plt.show()
#Showing mfcc_feat
plt.plot(mfcc)
plt.show()
```

```python
mfcc_data.shape
```

```python
melSpec_feat  = librosa.feature.melspectrogram(y=y_, sr=sr)
```

```python
melSpec_feat.shape
```

```python
ig, ax = plt.subplots()

cax = ax.imshow(melSpec_feat, interpolation='nearest', cmap=cm.coolwarm, origin='lower', aspect='auto')
ax.set_title('Mel_Spectrogram Features')
#Showing mfcc_data
plt.show()
#Showing mfcc_feat
plt.plot(mfcc)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
```

```python
librosa.display.specshow(librosa.power_to_db(melSpec_feat,ref=np.max),y_axis='mel', fmax=8000,x_axis='time')
```

```python
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()
```

```python
Spectral_bandwidth  = librosa.feature.spectral_bandwidth(y=y_, sr=sr)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.semilogy(Spectral_bandwidth.T,label='Spectral bandwidth')
plt.ylabel('Hz')
plt.xticks([])
plt.xlim([0, Spectral_bandwidth.shape[-1]])
plt.legend()
plt.show()
```
